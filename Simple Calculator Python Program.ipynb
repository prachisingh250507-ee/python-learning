```python
# Program to Create a simple calculator that can perform basic arithmetic
# operations such as addition, subtraction, multiplication, and division.
```

```python
num1=float(input("Enter the first number: "))
num2=float(input("Enter the second number: "))

print("FOR ADDITION CHOOSE 1\nFOR SUBTRACTION CHOOSE 2\nFOR MULTIPLICATION CHOOSE 3\nFOR DIVISION CHOOSE 4")

operation=int(input("Enter the operation you want to perform: "))

if operation==1:
  print("Addition of given numbers is: ", num1+num2)
elif operation==2:
  print("Subtraction of given numbers is: ", num1-num2)
elif operation==3:
  print("Multiplication of given numbers is: ", num1*num2)
elif operation==4:
  print("Division of given numbers is: ", num1/num2)
else:
  print("Invalid input")
```
